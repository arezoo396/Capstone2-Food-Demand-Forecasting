# Data wrangling

## 1-Imports

```python
#Import pandas
import pandas as pd
```

## 2-Load the data


```python
# Load The meal delivery company Data: Food Orders, Fulfilment center info, and meals info
orders = pd.read_csv('../Data/Food_orders_data.csv')
centers = pd.read_csv('../Data/fulfilment_center_info.csv')
meals= pd.read_csv('../Data/meal_info.csv')
```

## 3-Explore the data

```python
meals.head()
```

```python
#join all three tables
food_data=pd.merge(pd.merge(orders,centers, on='center_id'),meals, on='meal_id')
food_data.head()
```

```python
#change 'category' column name 
food_data.rename(columns={'category':'meal_category'},inplace=True)
```

### Distribution of feature values

Looking at distributions of features to get a feel for whether the values look sensible and whether there are any obvious outliers to investigate.

```python
#Use hist method to plot histograms of each of numeric features
food_data.hist(figsize=(15,10))
```

The only concern is the 'num_orders' variable because while most values are low, it has a maximum of 24,299. This could be an outlier, or it might represent a special week with an unusually high number of orders for a specific meal.

```python
# investigate num_orders values
# filter 'num_orders' column to print the values greater than 10,000
high_demand=food_data[food_data.num_orders>10000].sort_values('num_orders')
high_demand
```
